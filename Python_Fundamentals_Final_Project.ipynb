```python
#Import data related libraries
import pandas as pd
import numpy as np
import seaborn as sns

#Import GUI Libraries
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display

```

```python
#Load Car Crashes dataset from seaborn and display head information
df_car_crashes = sns.load_dataset("car_crashes")
df_car_crashes.head()
```

```python
#Statistical measures of crash data
df_car_crashes.describe()
```

```python
#Group By Alcohol and Speeding
df_car_crashes_group = df_car_crashes.groupby(["speeding", "ins_losses"])

for key, item in df_car_crashes_group:
    print(df_car_crashes_group.get_group(key), "\n\n")
```

```python
#create crosstab of data
this = pd.crosstab(columns = df_car_crashes["speeding"], index = [df_car_crashes["alcohol"], df_car_crashes["no_previous"], df_car_crashes["ins_premium"], df_car_crashes["ins_losses"]])

this
```

```python
#Create widgets to make the data stylish
#1. Sort Car crashes by state

ALL = "ALL"
def unique_crash_values_ALL(array):
  unique = array.unique().tolist()
  unique.sort()
  unique.insert(0, ALL)
  return unique
```

```python
#Create Dropdown to select state values
drop_downState = widgets.Dropdown(options = unique_crash_values_ALL(df_car_crashes.abbrev))
```

```python
def dropdown_state_handler(change):
  if(change.new == ALL):
    display(df_car_crashes)
  else:
    display(df_car_crashes[df_car_crashes.abbrev == change.new])
```

```python
drop_downState.observe(dropdown_state_handler,  names = "value")
```
